```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
celcius = np.array([-40,-10,0,8,15,22,38],dtype=float)
farenheit = np.array([-40,-14,32,46.4,59,71.6,100.4],dtype=float)
```

```python
capa= tf.keras.layers.Dense(units=1,input_shape=[1])
modelo = tf.keras.Sequential([capa])
```

```python
modelo.compile(
    optimizer = tf.keras.optimizers.Adam(0.1),
    loss = "mean_squared_error"
)
```

```python
print("comenzando entrenamiento .....")
historial = modelo.fit(celcius,farenheit,epochs=500,verbose=False)
print("modelo entrenado")
```

```python
plt.xlabel("# de iteracion")
plt.ylabel("magnitud de la perdida")
plt.plot(historial.history["loss"])
```

```python
print("realizar una prediccion")
resultado = modelo.predict([50])
print("resultado es "+ str(resultado) + "  farenheit")
```
